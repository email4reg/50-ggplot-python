```python
import pandas as pd
mpg= pd.read_csv("http://goo.gl/uEeRGu") 
```

```python
mpg.head()
```

```python
from altair import *
Chart(mpg).mark_bar(barThinSize=6).encode(x=X('displ'),
                            y=Y('count(*)'),
                             color='class'
                               ).configure_cell(width=500, height=200)
```

### Target 2- Fixed # bins

![](http://r-statistics.co/screenshots/ggplot_masterlist_20.png)

```python
from altair import *
Chart(mpg).mark_bar(barThinSize=6).encode(x=X('displ',bin=Bin(
            maxbins=5.0,
        )),
                            y=Y('count(*)'),
                             color='class'
                               ).configure_cell(width=500, height=200)
```
