```python
import pandas as pd
df= pd.read_csv("https://raw.githubusercontent.com/jkeirstead/r-slopegraph/master/cancer_survival_rates.csv") 
```

```python
df.head()
```

```python
from altair import *
#c = Chart(df).mark_line(color='#CD4F39').encode(x='year',y='group').configure_cell(width=200, height=200) +\ 
Chart(df).mark_text(color='#CD4F39').encode(x='year',
                                            y=Y('group', sort=SortField(op='mean', field='value', order='descending')),
                                            text='value').configure_cell(width=200, height=200) 
```

```python
Chart(df).mark_line(color='#CD4F39').encode(x='year',
                                            y=Y('group', sort=SortField(op='mean', field='value', order='descending'))).configure_cell(width=200, height=200) 
```
