```python
import pandas as pd
mpg= pd.read_csv("http://goo.gl/uEeRGu") 
```

```python
mpg.head()
```

```python
from altair import *
Chart(mpg).mark_bar(color='#CD4F39').encode(x=X('manufacturer', 
                                 title='Make',
                                sort=SortField(op='mean', order='ascending', field='cty')),
                            y=Y('mean(cty)', 
                                title='Mileage'
                               )).configure_cell(width=200, height=200)
```
