```python
import pandas as pd
mpg= pd.read_csv("http://goo.gl/uEeRGu") 
mlist = ["audi", "ford", "honda", "hyundai"]
mpg_select  = mpg.query('manufacturer==@mlist')
```

```python
from altair import *
chart = Chart(mpg_select)
chart.mark_circle().encode(
    x='displ',
    y='cty',
    size='hwy',
    color='manufacturer'
)
```
