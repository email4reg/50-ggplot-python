```python
from altair import *
Chart(mpg).mark_bar(barThinSize=6).encode(x=X('manufacturer', axis=Axis(labelAngle=-30)),
                            y=Y('count(*)'),
                             color='class'
                               ).configure_cell(width=500, height=200)
```
