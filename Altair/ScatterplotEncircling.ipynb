```python
import pandas as pd
midwest= pd.read_csv("http://goo.gl/G1K41K") 
```

```python
from altair import *
X()
chart = Chart(midwest)
chart.mark_circle().encode(
    x=X('area', title='Area'),
    y=Y('poptotal', title='Population'),
    color=Color('state'),
    size='popdensity',
).transform_data(filter = (expr.df.poptotal<50000))
```
